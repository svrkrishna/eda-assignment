```python
import requests
import pandas as pd
```

```python
# Task 1 — Fetch and Validate the Response

url = "https://restcountries.com/v3.1/all"

try:
    response = requests.get(url, timeout=10)
    print("HTTP Status Code:", response.status_code)
    
    if response.status_code == 200:
        countries_data = response.json()
        print("Total number of countries retrieved:", len(countries_data))
    else:
        print("Request failed. Could not retrieve country data.")
        
except requests.exceptions.RequestException as e:
    print("An error occurred while making the API request:", e)
```

```python
# Task 2 — Extract and Display Key Fields

country_list = []

for country in countries_data:
    common_name = country.get("name", {}).get("common", None)
    population = country.get("population", None)
    region = country.get("region", None)
    area = country.get("area", None)
    
    country_list.append({
        "common_name": common_name,
        "population": population,
        "region": region,
        "area": area
    })

df = pd.DataFrame(country_list)

df.head(10)
```

```python
df.info()
```

```python
# Task 3 — Summarize the Data

# Region with the highest total population
region_population = df.groupby("region")["population"].sum().sort_values(ascending=False)

highest_population_region = region_population.idxmax()
highest_population_value = region_population.max()

# Country with the largest area
largest_area_row = df.loc[df["area"].idxmax()]
largest_area_country = largest_area_row["common_name"]
largest_area_value = largest_area_row["area"]

print("Region with the highest total population:", highest_population_region)
print("Total population in that region:", highest_population_value)

print("\nCountry with the largest area:", largest_area_country)
print("Largest area:", largest_area_value)
```

## Summary

Using the REST Countries API, I fetched live country data and stored selected fields in a Pandas DataFrame.  
From the analysis, I identified the region with the highest total population and the country with the largest area.
